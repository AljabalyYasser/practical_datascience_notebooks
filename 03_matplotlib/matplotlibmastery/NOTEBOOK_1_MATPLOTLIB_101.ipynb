# Notebook 1: Matplotlib Basics

**Creating your first simple visualizations with Matplotlib**

## Why Visualize Numerical Data?

Suppose you track lemonade sales over several days:

- Monday: 50 cups
- Tuesday: 62 cups
- Wednesday: 55 cups

A list of raw numbers is informative, but a chart makes the pattern easier to understand. A plot helps you quickly identify:

- which day had the highest sales,
- whether sales are increasing or decreasing,
- and how values compare across time.

**Objective of this notebook:** create and understand a basic line plot.

```python
import matplotlib.pyplot as plt

# Magic for Jupyter notebooks (makes plots appear below cells)
%matplotlib inline
```

## Imports and Notebook Display

- `import matplotlib.pyplot as plt` imports Matplotlib's plotting interface.
- `%matplotlib inline` ensures that generated figures are displayed directly below the code cell in a Jupyter notebook.

## Data Preparation

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]  # X-axis
sales = [50, 62, 55, 73, 65]  # Y-axis
```

## Understanding the Dataset

- `days` contains the category labels shown on the **x-axis**.
- `sales` contains the numerical values shown on the **y-axis**.

Together, these two lists define the points that will be plotted.

## Basic Line Plot

```python
plt.plot(days, sales, color="purple", marker="o")
plt.title("Lemonade Sales This Week")
plt.xlabel("Day")
plt.ylabel("Cups Sold")
plt.show()
```

## Interpreting the Plot Code

1. `plt.plot(days, sales)` creates the line chart.
   - `color="purple"` sets the line color.
   - `marker="o"` adds a circular marker at each data point.
2. `plt.title(...)` adds a chart title.
3. `plt.xlabel(...)` and `plt.ylabel(...)` label the axes.
4. `plt.show()` renders the final figure.

## Plot Customization

```python
plt.plot(days, sales, color="#FF5733", marker="D", linestyle=":", linewidth=2)
plt.grid(True, alpha=0.3)
plt.title("Fancy Lemonade Sales", fontsize=14)
plt.show()
```

## Additional Formatting Options

- `color="#FF5733"` applies a custom hexadecimal color.
- `marker="D"` uses diamond-shaped markers.
- `linestyle=":"` changes the line to a dotted style.
- `linewidth=2` makes the line slightly thicker.
- `plt.grid(True, alpha=0.3)` adds a light background grid for readability.

These options improve presentation and make charts easier to interpret.

## Exercise: Ice Cream Shop Sales

**Given data**

- Days: `["Mon", "Wed", "Fri"]`
- Sales: `[120, 85, 110]`

**Task**

Create a line chart that:

1. uses **green star markers**, 
2. displays a **dashed line** (`"--"`),
3. has the title **"Weekly Ice Cream Sales"**,
4. and includes grid lines.

## Sample Solution

```python
days = ["Mon", "Wed", "Fri"]
sales = [120, 85, 110]

plt.plot(days, sales, color="green", marker="*", linestyle="--")
plt.title("Weekly Ice Cream Sales")
plt.grid(True)
plt.show()
```

## Key Takeaways

- Numerical data can be communicated more clearly through visualizations.
- Basic line plots can be customized with colors, markers, line styles, and grids.
- Clear titles and axis labels are essential for readable charts.
